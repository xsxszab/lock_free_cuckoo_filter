```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set_style('darkgrid')
```

```python
frame1 = pd.read_csv('data/ghc_thread_compare.csv')
```

```python
frame1
```

```python
fig, ax1 = plt.subplots(figsize=(5, 4), dpi=200)
plt.title("Throughput as A Function of Thread Count on GHC Machines")
sns.lineplot(data=frame1, x='Thread Count', y='Throughput (ops/sec)', hue='Type', ax=ax1, marker="o")
x_ticks = [2, 4, 8, 16, 32]
ax1.set_xticks(x_ticks)
plt.ticklabel_format(style='plain', axis='y')

plt.savefig('pic/ghc_compare.png')
```
